```python
#import library
import pandas as pd

#import CSV file
df = pd.read_csv('victims-of-selected-major-offences-by-age-group.csv',delimiter=',', encoding="utf-8-sig")
df
```

```python
#checking unique values in crime_type and age column
df.loc[:,'crime_type'].unique()
```

```python
df.loc[:, 'age'].unique()
```

```python
#removing non age and unknown age value in 'age' column
df = df[df.age != 'Unknown Victims Of Cheating & Related']
df = df[df.age != 'Unknown Age']
df
#only want victims under 21 --> remove 'above 21 years old' from age column
df = df[df.age != 'Above 21 Years Old']
df.head(20)
```

```python
#rename 'value' column to 'num_people' to reflect accuracy of dataset
df.rename(columns={"value": "num_people"}, inplace = True)
df
#export cleaned data as csv file
df.to_csv('cleaned_dataset.csv')
```

```python
#groupby total victims of crime_type by year then export as csv file
total_victims = df[df.crime_type.str.contains("Total Victims")]
total_victims.groupby(['year','crime_type'])[['num_people']].sum()
#total_victims.reset_index().to_csv('total_victims_only.csv')
```

```python
total_victims = total_victims.reset_index()
```

```python
#drop index column from total_victims
total_victims

group_total_victims = total_victims.groupby(['year','crime_type'])[['num_people']].sum()
```

```python
#total_victims.to_csv('total_victims_under21.csv', index=False)
group_total_victims
group_total_victims.to_csv('groupby_total_victims_under_21.csv')
group_total_victims
```

```python
import numpy as np
table = pd.pivot_table(group_total_victims, values='num_people', index=['year'],columns=['crime_type'], aggfunc=np.sum)
table
```

```python
#renaming the columns

table.columns = table.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
table
table.to_csv('pivot_total_victims_under21.csv')
```
